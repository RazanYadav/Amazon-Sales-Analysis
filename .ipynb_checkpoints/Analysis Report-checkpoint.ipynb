# __3. Exploratory Data Analysis with Data Visualization__ 

## __a. Basic statistics__

```python
print("Basic statistics:\n", data.describe())
```

## __b. Distribution of Product Prices__
Visualize the distribution of product prices to understand the price range and frequency.

```python
# Plot the histogram
plt.figure(figsize=(12,6))
plt.hist(data['Amount'], bins=30, color="violet", edgecolor='k', alpha=0.8)
plt.title('Sales Distribution Report')
plt.xlabel('Amount')  # Taking label from the dataset column
plt.ylabel('Frequency')
plt.show()
```

## c. Plotting Total Quantity per Category

```python
# Group by 'Category' and sum 'Qty'
category_review_count = data.groupby('Category')['Qty'].sum()
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 5))
plt.bar(category_review_count.index, category_review_count.values, color='lightgreen', edgecolor='k')
plt.title('Total Quantity per Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)
plt.show()
```

## d. Total Sales Distribution by Category


```python
# Group by 'Category' and sum the 'Amount' for total sales
total_sales_by_category = data.groupby('Category')['Amount'].sum()
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 5))
plt.bar(total_sales_by_category.index, total_sales_by_category.values, color="lightgreen", edgecolor='k')
plt.title('Sales Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

## e. Monthly Sales Analysis

```python
# Convert 'Date' column to datetime format with a specified format
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d')  # Adjust the format as needed

# Extract month and year for grouping
data['Month'] = data['Date'].dt.to_period('M')

# Calculate monthly totals
monthly_totals = data.groupby('Month')['Amount'].sum()
```

```python
# Plot the monthly sales
plt.figure(figsize=(12, 6))
plt.plot(monthly_totals.index.astype(str), monthly_totals.values, marker='o', color="green")
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

## f. Price vs. Quantity Analysis

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Amount'], data['Qty'], color="purple")  # Use 'Amount' for price and 'Qty' for review score
plt.title('Price vs. Quantity')
plt.xlabel('Price (Amount)')
plt.ylabel('Quantity')
plt.show()
```

## g. Sales Performance by Price Range


```python
# Create a copy of the dataset
data2 = data.copy()
```

```python
# Define price bins using 'Amount'
price_bins = [0, 50, 100, 200, 500, 1000]
data2['price_range'] = pd.cut(data2['Amount'], bins=price_bins)
```

```python
# Calculate total sales by price range using 'Qty'
sales_by_price_range = data2.groupby('price_range', observed=True)['Qty'].sum()
```

```python
# Plot the sales performance by price range
plt.figure(figsize=(10, 6))
sales_by_price_range.plot(kind='bar', color='lightgreen', edgecolor='k')
plt.title('Sales Performance by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
